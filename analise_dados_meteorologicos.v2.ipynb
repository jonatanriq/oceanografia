```python
import pandas as pd
import matplotlib.pyplot as plt
dados = pd.read_csv('meteorologia_rj.csv', sep=',')
```

```python
x = dados.groupby('Mes')['Nebulosidade'].mean()
y = dados.groupby('Mes')['Precipitacao'].mean()
plt.figure(figsize=(20,10))
plt.bar(dados['Mes'].unique(), y, color='green')
plt.xticks(range(0,13,1))
plt.xlabel('Meses', fontsize=12)
plt.ylabel('Precipitacao (mm)', fontsize=12)
plt.title('MÃ©dia Mensal de Chuva Para Toda SÃ©rie', fontsize=16)
plt.ylim([1,10])
plt.minorticks_on()
plt.tick_params()
plt.grid()
plt.show()
```

```python
# criar uma coluna condicional para valores de precipitacao
dados['Chuva'] = np.where(dados['Precipitacao']!= 0, 'Sim','NÃ£o')

#separando as estaÃ§Ãµes
dados_e1 = dados.query('Estacao == 83379.')
dados_e2 = dados.query('Estacao == 82024.')
dados_e3 = dados.query('Estacao == 83096.')
```

```python
fig, ax = plt.subplots(ncols=3, figsize=(20,10))

#definindo dados da primeira estaÃ§Ã£o
chuvas_e1 = dados_e1.query('Precipitacao > 0') 
chuva_e1 = chuvas_e1.groupby('Mes')['Chuva'].describe()

#definindo dados da segunda estaÃ§Ã£o
chuvas_e2 = dados_e2.query('Precipitacao > 0') 
chuva_e2 = chuvas_e2.groupby('Mes')['Chuva'].describe()

#definindo dados da terceira estaÃ§Ã£o
chuvas_e3 = dados_e3.query('Precipitacao > 0') 
chuva_e3 = chuvas_e3.groupby('Mes')['Chuva'].describe()



#plotando a primeira imagem
ax[0].bar(chuvas_e1['Mes'].unique(), chuva_e1['count'])
ax[0].set_xticks(range(1,13,1))
ax[0].set_title('EstaÃ§Ã£o N. 83379')
ax[0].set_ylabel('NÃºmero de OcorrÃªncias de Chuva', fontsize=14)
ax[0].grid()
ax[0].set_yticks(np.arange(0,850,50))

# plotando a segunda imagem
ax[1].bar(chuvas_e2['Mes'].unique(), chuva_e2['count'], color='green')
ax[1].set_xticks(range(1,13,1))
ax[1].set_title('EstaÃ§Ã£o N. 82024')
ax[1].set_ylabel('NÃºmero de OcorrÃªncias de Chuva', fontsize=14)
ax[1].grid()
ax[1].set_yticks(np.arange(0,850,50))



#plotando a terceira estaÃ§Ã£o
ax[2].bar(chuvas_e3['Mes'].unique(), chuva_e3['count'], color='grey')
ax[2].set_xticks(range(1,13,1))
ax[2].set_title('EstaÃ§Ã£o N. 83096')
ax[2].set_ylabel('NÃºmero de OcorrÃªncias de Chuva', fontsize=14)
ax[2].grid()
ax[2].set_yticks(np.arange(0,850,50))



plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib as mpl
```

```python
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['font.size'] = 14

dados1 = dados.iloc[::100,:]
```

```python


x1 = dados1['TempMinima']
y1 = dados1['TempBulboUmido']
x2 = dados1['TempMinima']
y2 = dados1['TempBulboSeco']


xmin = -10
xmax = 40
ymin = -10
ymax = 40


# cores 

colors = sns.color_palette("rocket",6)


# preparando o multipanel

fig = plt.figure(figsize=(5,5)) # tamanho da figura
gs = gridspec.GridSpec(4,4) # divide o grid
gs.update(wspace=0.2, hspace=0.25) #  as distÃ¢ncias entre os paineis

# gerando o primeiro painel

xtr_subsplot = fig.add_subplot(gs[0:4,0:4])


#plotando os dados para o primeiro painel

plt.scatter(x1,y1,color=colors[4],marker='o', label='Bulbo Ãšmido')

# definir onde voce quer os ticks

xticks = np.arange(-5,40,5)
yticks = np.arange(-10,40,5)

# parÃ¢metros dos ticks

plt.minorticks_on()
plt.tick_params(direction='in', which='minor', length=5, bottom=True, left=True)
plt.tick_params(direction='in', which='major', length=10, bottom=True, left=True, width=1.2)
plt.legend()
plt.grid()
plt.yticks(yticks)
plt.xticks(xticks)
plt.xlim(xmin,xmax)
plt.ylim(ymin,ymax)
plt.show()

# criando o segundo painel

xtr_subsplot = fig.add_subplot(gs[0:4,2:4])

plt.scatter(x2, y2, color=colors[2], marker='o', label='Bulbo Seco')
xticks = np.arange(-5,46,5)
yticks = np.arange(-10,46,5)

# parÃ¢metros dos ticks do segundo painel

plt.minorticks_on()
plt.tick_params(direction='in', which='minor', length=5, bottom=True, left=True)
plt.tick_params(direction='in', which='major', length=10, bottom=True, left=True, width=1.2)
plt.legend()
plt.grid()
plt.yticks(yticks)
plt.xticks(xticks)
plt.xlim(xmin,xmax)
plt.ylim(ymin,ymax)

plt.show()
```
