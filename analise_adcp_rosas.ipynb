```python
import pandas as pd
```

```python
data = pd.read_csv(diretorio, sep=';')
```

```python
temperatura = data['Atmp']
pressao = data['Pres']
tempo = pd.to_datetime(data['tempo'], format = '%d/%m/%Y %H:%M')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (20,10), dpi=300)
plt.plot(tempo,temperatura, '-b')
plt.xlabel('Tempo',fontsize=20)
plt.ylabel('Temperatura',fontsize=20)
plt.title('Série Temporal da Temperatura',fontsize=20)
plt.show()
```

```python
vel = ventos['Avg_Wnd_Sp']
direcao = ventos['Avg_Wnd_Dir_N']
new_labels = ["E", "NE", "N", "NW", "W", "SW", "S", "SE"]
ax = WindroseAxes.from_ax(theta_labels=new_labels)
ax.bar(direcao, vel, normed=True, opening=0.8, edgecolor='black')
ax.set_legend()
plt.title('Rosa Ventos', fontsize=20)
plt.show()
```
